```python
#Give a bar graph of most popular genres

genres = {}

@functools.lru_cache(maxsize=None)
def check_tag(tag, amount):
    if tag in genres:
        genres[tag] += amount

    else:
        genres[tag] = amount

for i in range(10):
    rec = dataset['tags_mb'][i]
    
    #print(dataset['tags_mb'][i])
    if type(rec) == str:

        rec_tags = rec.split(';')

        for i in range(len(rec_tags)):
            rec_tags[i] = rec_tags[i].strip
        
        print(rec_tags)
        
        for tag in rec_tags:
            check_tag(tag, dataset['listeners_lastfm'][i] / 10000)

print(genres)
        
```
